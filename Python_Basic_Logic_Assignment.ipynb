```python
# 1.User Login Check 
# Stored correct credentials information
correct_username = "admin"
correct_password = "1234"

# Taking username input from user
username = input("Enter username: ")

#Taking the password input from user 
password = input("Enter password: ")

# Checking if the username is correct 
if username == correct_username: 
    
    #If username is correct check the password 
    if password == correct_password:
        print("Login Successful")
    else:
        print("Invalid Password")
else:
    #If usernmae is wrong
    print("Invalid Username")
```

```python
#2.Pass / Fail Analyzer
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initializing counters for pass and fail students
pass_count = 0
fail_count = 0

# Looping 
for mark in marks:
    
    # To check if the student passed
    if mark >= 50:
        pass_count += 1
    
    # If not passed then failed 
    else:
        fail_count += 1

# Print the pass count 
print("Pass Students:", pass_count)
#Print the fail count
print("Fail Students:", fail_count)
#Print the total students 
print("Total Students:", len(marks))
```

```python
#3.Simple Data Cleaner
# List of names 
names = [" Alice ", "bob", " CHARLIE "]

# Creating an empty list to store cleaned names
cleaned_names = []

# Looping through each name
for name in names:
    
    # Removes extra spaces
    name = name.strip()
    
    # Converts name to lowercase
    name = name.lower()
    
    # Adds cleaned name to new list
    cleaned_names.append(name)

# Prints the final cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
#4.Message Length Analyzer
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Looping through each message
for message in messages:
    
    # Calculating the message length
    length = len(message)
    
    # Printing message along with its length
    print("Message:", message)
    print("Length:", length)
    
    # Checking if message is longer than 10 characters
    if length > 10:
        print("This message is longer than 10 characters")
    print()
```

```python
#5.Error Message Detector
# List 
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counter to track number of errors
error_count = 0

# Looping through each log entry
for log in logs:
    
    # Checking if log entry is an error
    if log == "ERROR":
        error_count += 1   # Increasing the error counter

# Printing total number of error messages
print("Total ERROR messages:", error_count)

```
